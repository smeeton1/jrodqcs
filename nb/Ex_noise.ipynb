This example show how to use the noise that is uniform over all gates.
This file need the Plots module to run.

```julia
include("../src/myqcs.jl");
using Plots;
```

Following is an example of computing the fidelity curve for dephasing noise.

```julia
F=[];
push!(F,[0.1,0.125,0.15,0.175,0.2,0.25,0.3,0.4,0.5,0.6,0.7,0.75,0.8,0.825,0.85,0.875,0.9],[])
T=myqcs.noteb.qc_network(5,"Density",false,"line");
myqcs.proform.Random_Circuit(T,15,10);
myqcs.tensor_fun.simple_solver(T);
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.1);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.125);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.15);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.175);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.2);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.25);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.3);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.4);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.5);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.6);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.7);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.75);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.8);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.825);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.85);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.875);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
TE=deepcopy(T);
myqcs.noise.add_nosie(TE,0.9);
myqcs.tensor_fun.simple_solver(TE);
push!(F[2],myqcs.proform.fidelity(TE.out[1],TE.out[2]));
a = plot(F[1],F[2], fmt = :png,legend=false,linecolor="black",size = (500, 300), grid=false)
```
